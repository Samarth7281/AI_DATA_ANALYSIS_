```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 18, 16]  # For line plot
y2 = [3, 7, 5, 9, 6]       # For bar plot

# Create subplots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Line plot in first subplot
ax1.plot(x, y1, color='blue', marker='o')
ax1.set_title('Line Plot')
ax1.set_xlabel('X')
ax1.set_ylabel('Y1')

# Bar plot in second subplot
ax2.bar(x, y2, color='orange')
ax2.set_title('Bar Plot')
ax2.set_xlabel('X')
ax2.set_ylabel('Y2')

# Improve layout spacing
plt.tight_layout()

# Show the combined figure
plt.show()
```
