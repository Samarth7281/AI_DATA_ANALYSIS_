```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]

# Create a styled line plot
plt.plot(x, y, color='darkorange', linestyle='--', marker='o', linewidth=2, markersize=8, label='Data Line')

# Add title and axis labels
plt.title('Styled Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show legend
plt.legend()

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```
