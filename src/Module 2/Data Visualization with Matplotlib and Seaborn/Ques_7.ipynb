```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    'Height': [160, 170, 165, 180, 175],
    'Weight': [55, 65, 60, 80, 76],
    'Size': [10, 15, 13, 18, 12]  # Custom size for each point
}

# Create DataFrame
df = pd.DataFrame(data)

# Create scatter plot with Seaborn
sns.scatterplot(
    data=df,
    x='Height',
    y='Weight',
    size='Size',
    sizes=(50, 300),        # Scale of point sizes
    legend=False,
    color='purple'
)

# Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight with Size Encoding')

# Show plot
plt.show()
```
