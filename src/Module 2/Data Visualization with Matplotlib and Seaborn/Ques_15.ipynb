```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset('iris')

# Compute correlation matrix (excluding categorical column)
corr_matrix = iris.drop(columns='species').corr()

# Create heatmap with advanced styling
sns.set(style='white')  # Clean white background
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,           # Show correlation coefficients
    cmap='coolwarm',      # Color scheme
    fmt=".2f",            # Format for annotations
    linewidths=0.5,       # Lines between cells
    square=True,          # Keep squares
    cbar_kws={"shrink": 0.8}  # Colorbar size
)

# Add title
plt.title('Correlation Matrix of Iris Features', fontsize=14)
plt.tight_layout()
plt.show()
```
