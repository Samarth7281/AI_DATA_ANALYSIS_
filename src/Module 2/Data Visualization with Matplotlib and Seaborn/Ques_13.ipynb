```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create customized boxplot
sns.boxplot(
    data=iris,
    x='species',
    y='petal_length',
    palette='pastel',
    width=0.6,
    linewidth=2
)

# Add labels and title
plt.title('Petal Length Distribution by Iris Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Show the plot
plt.show()
```
